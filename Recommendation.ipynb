```python
import pandas as pd
df = pd.read_csv('cleaned_data_amazon.csv')
```

```python
pivot_table = df.pivot_table(index='user_id', columns='product_id', values='rating').fillna(0)
corr_matrix = pivot_table.corr(method='pearson')
```

```python
def recommend_similar_products(product_id, corr_matrix, top_n=3):
    if product_id not in corr_matrix.columns:
        print(f"Product ID {product_id} not found in data.")
        return -1
    similar_scores = corr_matrix[product_id].drop(labels=[product_id])
    top_similar = similar_scores.sort_values(ascending=False).head(top_n)
    return list(top_similar.index)
```

```python
product_to_recommend = 'B07JW9H4J1'
recommendations = recommend_similar_products(product_to_recommend, corr_matrix, top_n=3)

print("User-Item Rating Matrix:")
print(pivot_table)
print("\nProduct-Product Correlation Matrix:")
print(corr_matrix)
print(f"\nTop recommendations similar to product {product_to_recommend}:")
print(recommendations)
```
