```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv('amazon.csv')
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df1['discounted_price'] = df1['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df1['actual_price'] = df1['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
df1['discount_percentage'] = df1['discount_percentage'].str.replace('%','').astype('float64')

df1['discount_percentage'] = df1['discount_percentage'] / 100
```

```python
df1['rating']=df1['rating'].str.replace('|','2').astype(float)
```

```python
df1['rating_count']=df1['rating_count'].str.replace(',', '').astype(float)
```

```python
df1.describe()
```

```python
df1.head()
```

```python
df1['rating_count'].fillna(df1['rating_count'].mean(), inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1.to_csv('cleaned_data.csv')
```
